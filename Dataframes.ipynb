```python
# 1 Copy the code from the lesson to create a dataframe full of student grades.

import pandas as pd
import numpy as np

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

```python
# a. Create a column named passing_english that indicates whether each student has a passing grade in english.
df = df.assign(passing_english = df.english >= 70 )
```

```python
#### alternative

df['passing_english'] = df['english'] >= 70
```

```python
df.head()
```

```python
# b. Sort the english grades by the passing_english column. How are duplicates handled?
#duplicates are sorted by indes
```

```python
df.sort_values(by='passing_english')
```

```python
# c Sort the english grades first by passing_english and then by student name.
# All the students that are failing english should be first, and within the students that are failing 
# english they should be ordered alphabetically. The same should be true for the students passing english. (Hint: you can pass a list to the .sort_values method)
df.sort_values(by= 'name').sort_values(by = 'passing_english')
```

```python
#### alternative
df.sort_values(by=['passing_english','name'])
```

```python
# d Sort the english grades first by passing_english, and then by the actual english grade, 
# similar to how we did in the last step.
```

```python
df.sort_values(by= 'passing_english').sort_values(by = 'english')
```

```python
df.sort_values(by=['passing_english', 'english'], ascending = [True, False])
```

```python
# e Calculate each students overall grade and add it as a column on the dataframe.
# The overall grade is the average of the math, english, and reading grades.
```

```python
df = df.assign(overall_grade = (df.english + df.math + df.reading)/3).round(2)
df
```

```python
df['overall']= ((df.math + df.english + df.reading)/3).round()
```

```python
# Do any cars have better city mileage than highway mileage?
    # no car in the list has better city milage than highway milage
bool_mask = (mpg['city']) > (mpg['highway'])
mpg[bool_mask]
```

```python
# Create a column named mileage_difference this column should contain the difference between highway 
# and city mileage for each car.
mpg = mpg.assign(milage_difference= (mpg['highway']) - (mpg['city']))  
```

```python
# Which compact class car has the lowest highway mileage? The best?
compact_cars = mpg[mpg['class']=='compact']
compact_cars
```

```python
# the worst milage
bool_mask= compact_cars['highway']== compact_cars['highway'].min()
compact_cars[bool_mask]
```

```python
# the best milage
bool_mask= compact_cars['highway']== compact_cars['highway'].max()
compact_cars[bool_mask]
```

```python
# Create a column named average_mileage that is the mean of the city and highway mileage.
mpg = mpg.assign(average_milage = (mpg.city + mpg.highway)/2)
mpg.head()
```
