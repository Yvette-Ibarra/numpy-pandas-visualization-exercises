## 3. Would it be reasonable to predict species based on sepal width and sepal length? For this, you'll visualize two numeric columns through the lense of a categorical column.

####  Sepal width and sepal length are not good indicators of  the species

```python
sns.relplot(data = iris, x = 'sepal_length', y = 'sepal_width', hue ='species')
```

## 4. Which features would be best used to predict species?

#### Petal length would be the best feature to predict species next best features is petal width.

```python
# Create a figure with 4 boxplots
plt.figure(figsize = (10,8))

# first subplot
plt.subplot(221)
sns.boxplot(data = iris, y = 'sepal_width', x= 'species')

# second subplot
plt.subplot(222)
sns.boxplot(data = iris, y = 'sepal_length', x= 'species')

# third subplot
plt.subplot(223)
sns.boxplot(data = iris, y = 'petal_width', x= 'species')

# fouth subplot
plt.subplot(224)
sns.boxplot(data = iris, y = 'petal_length', x= 'species')
```

```python
sns.pairplot(iris, hue ='species',corner = True)
```
