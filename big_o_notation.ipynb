# Big O Notation

```python
import matplotlib.pyplot as plt
from random import randint
import math
import random
import numpy as np
```

```python
plt.figure(figsize = (8,8))

plt.title( '$Big O Notation$',c="mediumblue",fontsize= 36)
plt.xlabel('$Elements$')
plt.ylabel('$Operations$', rotation = 0)


x1 = list(range(1000))
y1 =[0+1 for i in x1]
plt.plot(x1, y1, color = 'green',label = '$y = O(1)$')


x2 =list([i for i in range(0,1000) if i != 0])
y2 = np.log(x2)
plt.plot(x2, y2, color = 'orange',label = '$y = O(log n)$')


x3 = list(range(1000))
y3 = x3
plt.plot(x3, y3, color = 'red',ls = "--",label = '$y = O(n)$')

x4 = np.array(range(0, 1001))
y4 = x4*np.log(x4)
plt.plot(x4, y4)
plt.plot(x4, y4, color = 'gray',ls = "--",label = '$y =O(nlog n)$')

x5 =  np.array(range(0, 10))
y5 = x5**2
plt.plot(x5, y5, color = 'pink',label = '$y = O(n^2)$')



x6 =  np.array(range(0, 1001))
y6 =np.power(2, x6)
plt.plot(x6, y6, color = 'purple',label = '$y = O(2^n)$')

x7 = list(range(0, 10))
y7 = [math.factorial(n) for n in x7]
plt.plot(x7, y7, color = 'seagreen',label = '$y = O(n!)$')

x8 = np.array(range(0, 10))
y8 = np.power(x8, x8)
plt.plot(x8, y8, color = 'lightblue',label = '$y = O(n^n)$')

plt.xlim(0, 50)
plt.ylim(0, 50)

plt.legend()
plt.show()
```
