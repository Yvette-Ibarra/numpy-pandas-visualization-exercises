
1. Use matplotlib to plot the following equation:

y= x^2 - x +2
You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.


```python
import matplotlib.pyplot as plt
from random import randint
import math
import random
```

```python
plt.figure(figsize = (6,5))
x = list(range(-25,25))
y = [n**2 -n + 2 for n in x]
plt.scatter(x, y)
plt.title( 'Quadratic',c="purple",fontsize= 36)
plt.xlabel('$x$')
plt.ylabel('$y$', rotation = 0)
plt.annotate('Origin', xy=(0, 0), xytext=(0, 50),arrowprops={'facecolor': 'gold'}, fontsize = 15)
plt.show()
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

 y= sqrt(x)
 y = x^3
 y = 2^x
 y = 1/(x+1)

You can use functions from the math module to help implement some of the equations above.

```python
plt.figure(figsize = (6,5))
x = list(range(0,25))
y = [n**(1/2) for n in x]
plt.scatter(x, y, color = 'seagreen', ls = '--', marker = '*')
plt.title( 'Square Root',c="Red",fontsize = 25)
plt.xlabel('$Input$')
plt.ylabel('$Output$')

plt.show()
```

```python
plt.figure(figsize = (6,5))
x = list(range(-25,25))
y = [n**(3) for n in x]
plt.scatter(x, y, color = 'green', ls = '--', marker = '8')
plt.title( 'Cube Root',c="gold",fontsize = 25)
plt.xlabel('$Independent$')
plt.ylabel('$Dependent$')

plt.show()
```

```python
plt.figure(figsize = (6,5))
x = list(range(-1,25))
y = [2**(n) for n in x]
plt.scatter(x, y, color = 'k', ls = '--', marker = '8')
plt.title( 'Exponential',c="teal",fontsize = 25)
plt.xlabel('$Independent$')
plt.ylabel('$Dependent$')

plt.show()
```

```python
plt.figure(figsize = (6,5))
x = list([i for i in range(-25,25) if i != -1])
y = [1/(n+1) for n in x]
plt.scatter(x, y, color = 'magenta', ls = '--', marker = '+')
plt.title( 'A Rational Scatter Plot',c="Indigo",fontsize = 25)
plt.xlabel('$x$')
plt.ylabel('$y$', rotation = 0)

plt.show()
```
