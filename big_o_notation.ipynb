```python
import matplotlib.pyplot as plt
from random import randint
import math
import random
import numpy as np
```

```python
plt.figure(figsize = (8,8))

plt.title( '$Big O Notation$',c="mediumblue",fontsize= 36)
plt.xlabel('$Elements$')
plt.ylabel('$Operations$', rotation = 0)


x1 = list(range(1000))
y1 =[i+1 for i in x1]
plt.plot(x1, y1, color = 'green',label = '$y = On + 1$')


x2 =list([i for i in range(0,1000) if i != 0])
y2 = np.log(x2)
plt.plot(x2, y2, color = 'orange',label = '$y = O(log n)"$')


x3 = list(range(1000))
y3 = x3
plt.plot(x3, y3, color = 'red',ls = "--",label = '$y = O(n)"$')

x4 = list([i for i in range(0,1000) if i != 0])
y4 = [x2*(math.log(i)) for i in x4]
plt.plot(x4, y4, color = 'red',ls = "--",label = '$O(n log n)"$')



plt.legend()
plt.show()
```
